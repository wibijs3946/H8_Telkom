### Data Cleaning

Dari hasil overview tentang nilai di setiap kolom tersebut, dapat diambil kesimpulan bahwa banyak data yang tidak valid karena "JML Peserta Hadir" ada yang bernilai 0. Data ini menjadi tidak valid karena akan mempengaruhi nilai yang lainnya. Jumlah data yang tidak valid dapat kita grouping berdasarkan academy-nya sebagai berikut:

```python
data_feed_pelatihan_fetch.loc[data_feed_pelatihan_fetch['JML Peserta Hadir'] == 0, 'Academy Event'].value_counts()
```

Pertama, kita akan membuang data dari "KMCS", "GOFM", dan "PNC, LO, dan GS" karena unit-unit tersebut tidak termasuk kedalam penyelenggara pelatihan sehingga seharusnya tidak memiliki nilai NPS dan dapat kita abaikan

```python
data_feed_pelatihan_fetch.drop(data_feed_pelatihan_fetch.loc[data_feed_pelatihan_fetch['Academy Event'] == 'KMCS'].index, inplace=True)
data_feed_pelatihan_fetch.drop(data_feed_pelatihan_fetch.loc[data_feed_pelatihan_fetch['Academy Event'] == 'GOFM'].index, inplace=True)
data_feed_pelatihan_fetch.drop(data_feed_pelatihan_fetch.loc[data_feed_pelatihan_fetch['Academy Event'] == 'PNC,LO, dan GS'].index, inplace=True)
```

Selanjutnya kita filter data dengan membuang rows yang berisi "JML Peserta Hadir" dengan nilai 0

```python
df_clean = data_feed_pelatihan_fetch[data_feed_pelatihan_fetch['JML Peserta Hadir'] > 0]
```

```python
df_clean.shape
```

```python
df_clean.describe()
```

Dari data tersebut, ternyata masih terdapat nilai 0 pada bagian UBPP sehingga menyebabkan nilai Customer Satisfaction, First Response & Average Handling, dan Customer Effort Score bernilai 0. Hal ini berarti peserta mengikuti pelatihan tetapi tidak mengisi Umpan Balik atau UBPP. Oleh karena itu data perlu kita filter sekali lagi untuk mengambil pelatihan yang ada nilai UBPP-nya saja

```python
df_clean = df_clean[df_clean['Customer Satisfaction'] > 0]
```

```python
df_clean.shape
```

Setelah data sudah bersih, data siap dilakukan processing lebih lanjut. Preprocessing selanjutnya kita akan merapikan data tanggal pelatihan. Pada bagian ini, kita akan menambahkan kolom bulan pada data tersebut berdasarkan tanggal selesai pelatihan sehingga data dapat diubah menjadi time series berdasarkan bulan. Kita akan melakukan pengecekan terhadap data tanggal selesai, apakah terdapat duplicate value berdasarkan tanggal selesai

```python
df_clean[df_clean.duplicated(['Tgl Selesai'])].shape
```

Dari hasil pengecekan tersebut, terdapat 692 data yang duplicated

```python
# Membuat 1 dataframe dengan mengambil value bulan dari Tgl Selesai
df_clean_bln = df_clean['Tgl Selesai'].apply(lambda x: x[5:7])
df_clean_bln = df_clean_bln.astype(int)


# Menambahkan kolom Bln Selesai
df_clean.insert(12, 'Bln Selesai', df_clean_bln)
```

```python
df_clean.info()
```

```python
# Reset index mulai dari 0
df_clean.index = range(len(df_clean.index))
df_clean.head(5)
```

## Area Plot

```python
# import library yang dibutuhkan
import matplotlib as mpl
import matplotlib.pyplot as plt
```

Kita akan melakukan Area Plot untuk menampilkan data NPS rata-rata di setiap bulan berdasarkan academy tertentu. Kita perlu membuat dataframe baru yang berisikan Academy dan nilai rata-rata NPS di setiap bulannya

```python
# Buat series bulan 1-12 yang akan digunakan dalam membuat dataframe selanjutnya
bln = list(map(int, range(1, 13)))
bln
```

```python
# Buat dataframe df_nps yang berisi data bulan, academy, dan nilai NPS dari setiap pelatihan dari dataframe df_clean
df_nps = pd.DataFrame({'Bln': df_clean['Bln Selesai'], 'Academy': df_clean['Academy Event'], 'NPS': df_clean['Net Promotor Score']})
```

```python
df_nps
```

```python
# Hitung rata-rata NPS per bulan per academy
df_nps = df_nps.groupby(['Bln','Academy']).mean().reset_index()
df_nps
```

```python
# Buat dataframe baru yaitu df_nps_plot dengan bulan 1-12 sebagai kolom
df_nps_plot = pd.DataFrame({'Academy': df_nps['Academy'].unique()})
df_nps_plot

# Tambahkan kolom bulan 1 sampai 12 di df_nps_plot sebagai empty value
for a in bln:
    df_nps_plot[a] = ''

# Isi cell di masing
for academy in df_nps_plot['Academy']:
    for a in bln:
        try:
            df_nps_plot.loc[df_nps_plot['Academy'] == academy, a] = df_nps.loc[(df_nps['Academy'] == academy) & (df_nps['Bln'] == a), 'NPS'].values[0]
        except IndexError:
            pass

# Set Academy sebagai index
df_nps_plot.set_index('Academy', inplace=True)
```

```python
df_nps_plot
```

```python
df_nps_plot.info()
```

Dari data df_nps_plot, ternyata ada missing value yaitu data NPS Academy Leadership di bulan 4 dan Academy WS & International di bulan 3. Selain itu, data nilai NPS ternyata dianggap sebagai object sehingga kita perlu melakukan konversi ke numeric values terlebih dahulu. 

Untuk dataset yang besar dan tidak dapat terlihat secara langsung, kita dapat menggunakan command berikut

```python
df_nps_plot = df_nps_plot.apply(pd.to_numeric)
df_nps_plot.info()
```

Setelah dataframe df_nps_plot berhasil dikonversi menjadi numeric, terlihat missing values di bulan 3 dan bulan 4. Selanjutnya kita akan mengisi missing values tersebut dari bulan sebelumnya. Langkah ini dilakukan dengan asumsi ketika tidak ada pelatihan, maka nilai NPS pada bulan tersebut sama seperti sebelumnya. 

```python
# Mengisi nilai NA di bulan 3 dan bulan 4 berdasarkan nilai NPS bulan sebelumnya
df_nps_plot[3].fillna(df_nps_plot[2], inplace=True)
df_nps_plot[4].fillna(df_nps_plot[3], inplace=True)
```

```python
df_nps_plot
```

Setelah data sudah siap untuk diplot, kita akan menggunakan dataframe baru untuk setiap chart agar data tidak tercampur

```python
# Transpose data agar dapat diplot ke Area Plot
df_nps_area_plot = df_nps_plot[bln].transpose()

# Ubah index value dari df_nps_area_plot menjadi integer untuk keperluan plotting
df_nps_area_plot.index = df_nps_area_plot.index.map(int)
```

```python
# Plotting ke area plot
df_nps_area_plot.plot(kind='area',
                      alpha=0.2,
                      stacked=True,
                      figsize=(20,10),                     
                     )

plt.title('Rata-rata Nilai NPS Academy per Bulan')
plt.ylabel('Nilai NPS')
plt.xlabel('Bulan')

plt.show()
```

Dari hasil area plot data NPS masing-masing Academy per bulan tersebut, terlihat bahwa hampir seluruh Academy mengalami penurunan nilai NPS kecuali Academy DSP dan Leadership. Hal ini terjadi karena pada Bulan Maret, Covid-19 outbreak mulai terjadi di Indonesia, sehingga Telkom Corpu mulai mengubah sistem pelatihan menjadi online pada Bulan April. Perubahan delivery pelatihan dari in-class menjadi online mempengaruhi nilai NPS cukup signifikan.

## Word Clouds

Pada bagian ini, kita akan mencoba melakukan visualisasi untuk istilah apa yang paling sering muncul dalam judul pelatihan yang diselenggarakan Telkom Corpu selama Tahun 2020.

```python
# Import packages wordcloud dan stopwords
from wordcloud import WordCloud, STOPWORDS
```

```python
# Kita akan mengambil judul-judul pelatihan dari df_clean
judul_pelatihan = ''.join(df_clean['Nama Pelatihan'])
judul_pelatihan
```

```python
# set STOPWORDS dari package wordcloud sebagai stopwords yang akan digunakan
stopwords = set(STOPWORDS)

# generate wordcloud
judul_wc = WordCloud(
    background_color='white',
    stopwords=stopwords
)

# menambahkan stopwords
stopwords.add('Batch')
stopwords.add('Class')

judul_wc.generate(judul_pelatihan)
```

```python
# Set ukuran wordcloud
fig = plt.figure()
fig.set_figwidth(14)
fig.set_figheight(18)

# Tampilkan wordcloud
plt.imshow(judul_wc, interpolation='bilinear')
plt.axis('off')
plt.show()
```

Dari analisis wordcloud terhadap judul pelatihan selama Tahun 2020, ternyata kata-kata yang sering muncul adalah Business, Hack Idea, Digital, Force (dari kata Sales Force), TR (Telkom Regional), dan Certified (sertifikasi). Hal ini selaras dengan tujuan perusahaan dimana pelatihan-pelatihan yang diselenggarakan sebagian besar berhubungan dengan bisnis dan digital.
